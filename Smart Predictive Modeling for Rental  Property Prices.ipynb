## Problem Statement:
  In the real estate industry, determining the appropriate rental price for a property is crucial for property owners, tenants, and property management companies. Accurate rent predictions can help landlords set competitive prices, tenants make informed rental decisions, and property management companies optimize their portfolio management.
  The goal of this project is to develop a data-driven model that predicts the rental price of residential properties based on relevant features. By analyzing historical rental data and property attributes, the model aims to provide accurate and reliable rent predictions.

1. **id**: 
    - This is a unique identifier assigned to each property listing. It helps distinguish one listing from another and is useful for data management and referencing.

2. **type**: 
    - Denotes the type of property being listed, such as BHK1 (1 Bedroom, Hall, Kitchen), BHK2 (2 Bedroom, Hall, Kitchen), RK1 (Room, Kitchen), etc. This field provides information about the basic layout or configuration of the property.

3. **locality**: 
    - Refers to the specific neighborhood or area where the property is situated. The locality can significantly impact various aspects of the property, including its desirability, amenities, and rental price.

4. **activation_date**: 
    - Indicates the date when the property listing became active or was made available for rent. This information is crucial for tracking the timeline of property listings and analyzing trends over time.

5. **latitude** and **longitude**: 
    - These are geographic coordinates representing the precise location of the property on the Earth's surface. Latitude specifies the north-south position, while longitude indicates the east-west position. Together, they provide the property's exact geographical reference point.

6. **lease_type**: 
    - Specifies the type of lease associated with the property, such as FAMILY, BACHELOR, or ANYONE. This attribute defines the intended or preferred tenant demographic for the property.

7. **gym**, **lift**, **swimming_pool**: 
    - These binary indicators denote whether the property offers specific amenities like a gym, lift (elevator), or swimming pool. These amenities can significantly influence the attractiveness and rental value of the property.

8. **negotiable**: 
    - Indicates whether the rental price for the property is negotiable. This attribute provides insight into the flexibility of pricing terms offered by the property owner or manager.

9. **furnishing**: 
    - Describes the level of furnishing provided in the property, such as fully furnished, partially furnished, or unfurnished. Furnishing status can impact tenant convenience, comfort, and rental rates.

10. **parking**: 
    - Specifies the availability of parking facilities associated with the property. This attribute informs potential tenants whether parking space is provided and may influence their decision-making process.

11. **property_size**: 
    - Represents the size or area of the property, typically measured in square footage or square meters. Property size is a fundamental characteristic affecting its utility, livability, and rental value.

12. **property_age**: 
    - Indicates the age of the property since its construction. Property age can influence maintenance requirements, structural integrity, and perceived value.

13. **bathroom**: 
    - Denotes the number of bathrooms available within the property. The bathroom count is a key factor in determining the property's suitability for tenants with specific occupancy needs.

14. **facing**: 
    - Specifies the direction in which the property faces, such as north, south, east, or west. Facing orientation can affect natural lighting, ventilation, and overall ambiance within the property.

15. **cup_board**: 
    - Indicates the presence of cupboards or storage units within the property. Adequate storage space is an essential consideration for tenants and can impact their satisfaction and convenience.

16. **floor** and **total_floor**: 
    - These attributes describe the floor number of the property within the building and the total number of floors in the building, respectively. They provide information about the property's vertical positioning and scale within its architectural context.

17. **amenities**: 
    - Represents additional amenities or features provided with the property beyond those explicitly mentioned earlier. This attribute may include features like security systems, recreational facilities, or landscaped gardens.

18. **water_supply**: 
    - Describes the type and availability of water supply serving the property. Reliable water supply is a critical utility for any residential property and can influence tenant comfort and satisfaction.

19. **building_type**: 
    - Specifies the architectural style or type of building to which the property belongs, such as Apartment or Individual House. This attribute characterizes the structural and functional aspects of the property within its broader context.

20. **balconies**: 
    - Indicates the number of balconies or outdoor spaces associated with the property. Balconies can enhance outdoor living experiences and provide additional recreational or relaxation areas for tenants.

21. **rent**: 
    - This is the target variable representing the rental price for the property. It is the primary focus of analysis and prediction in rental property datasets, influencing various stakeholders' decision-making processes.

# Regression Model Building

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
from sklearn.svm import SVR
from sklearn import metrics
```

```python
x = df_con.drop('Rent_log',axis=1)
y = df_con['Rent_log']
```

```python
x
```

```python
y
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.20, random_state=42)
```

```python
x_train
```

```python
x_train[0]
```

```python
y_train
```

```python
y_train[1]
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
rf =  RandomForestRegressor()
dtr = DecisionTreeRegressor()
knn = KNeighborsRegressor()
xg = XGBRegressor()
lr =  LinearRegression()

models = [rf, dtr, knn,xg]
metrics_list = []

for model in models:
    model.fit(x_train, y_train)
    y_train_pred = model.predict(x_train)
    y_pred = model.predict(x_test)
    r2_test = metrics.r2_score(y_test, y_pred)
    r2_train = metrics.r2_score(y_train, y_train_pred)

    model_metrics = {'Algorithm': str(model).split('(')[0],
                     'R2_train': r2_train,
                     'R2_test': r2_test}
                     

    metrics_list.append(model_metrics)
```

```python
best_models = pd.DataFrame(metrics_list)
best_models  
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid = {
    'n_estimators': [50, 100, 200],
    'max_features': [None, 'sqrt', 'log2'],
    'max_depth': [None, 10, 20],
    'min_samples_split': [2, 4, 8],
}

rf = RandomForestRegressor()
x_train, x_test, y_train, y_test = train_test_split( x, y, test_size=0.20, random_state=42)

grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5 , n_jobs=-1, verbose=3)
grid_search.fit(x_train, y_train)


print("Best Cross-validated Parameters: ", grid_search.best_params_)
print("Best Cross-validated Score: ", grid_search.best_score_)
```

```python
rf_rg = RandomForestRegressor(n_estimators = 200, min_samples_split = 4, max_depth = None, max_features = None)
rf_rg.fit(x_train,y_train)
```
